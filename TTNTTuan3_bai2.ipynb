Hàm minimax được triển khai nhằm tìm nước đi tối ưu cho AI trong trò chơi Tic-Tac-Toe.
Thuật toán giả định rằng cả AI và đối thủ đều chơi tối ưu, và lựa chọn nước đi dựa trên việc tối đa hóa lợi ích cho AI hoặc tối thiểu hóa lợi ích cho đối thủ.

**Tham số**

position :	Danh sách 1D 9 ô biểu diễn trạng thái bàn cờ hiện tại.

depth :	Độ sâu hiện tại trong cây tìm kiếm. Dùng để ưu tiên chiến thắng sớm và kéo dài thất bại.

alpha :	Giá trị tốt nhất mà người chơi MAX có thể đảm bảo tại nhánh hiện tại.

beta :	Giá trị tốt nhất mà người chơi MIN có thể đảm bảo tại nhánh hiện tại.

isMaximizing :True nếu lượt MAX (thường là X), False nếu lượt MIN (thường là O).
Kiểm tra trạng thái kết thúc (Base Case):

 **Nguyên lý hoạt động**

Nếu X thắng → trả về 10 - depth.

Nếu O thắng → trả về -10 + depth.

Nếu hòa → trả về 0.

Ý nghĩa: depth được trừ/cộng để ưu tiên chiến thắng nhanh hoặc kéo dài thất bại.

Lượt MAX (X):

Khởi tạo giá trị maxEval = -∞.

Thử tất cả các ô còn trống, đặt X vào ô và gọi đệ quy Minimax cho lượt MIN.

Cập nhật maxEval = max(maxEval, Eval) và alpha = max(alpha, Eval).

Hoàn tác nước đi (backtracking).

Nếu beta <= alpha → cắt tỉa nhánh (Alpha-Beta pruning).

Lượt MIN (O):

Khởi tạo giá trị minEval = +∞.

Thử tất cả các ô còn trống, đặt O vào ô và gọi đệ quy Minimax cho lượt MAX.

Cập nhật minEval = min(minEval, Eval) và beta = min(beta, Eval).

Hoàn tác nước đi (backtracking).

Nếu beta <= alpha → cắt nhánh không cần thiết.

````python

def minimax(position, depth, alpha, beta, isMaximizing):
    winner = GetWinner(position)
    if winner != None:
        return 10 - depth if winner == "X" else -10 + depth
    if len(GetAvailableCells(position)) == 0:
        return 0
    if isMaximizing:
        maxEval = -math.inf
        for cell in GetAvailableCells(position):
            position[cell - 1] = "X"
            Eval = minimax(position, depth + 1, alpha, beta, False)
            maxEval = max(maxEval, Eval)
            alpha = max(alpha, Eval)
            position[cell - 1] = cell
            if beta <= alpha:
                break
        return maxEval
    else:
        minEval = +math.inf
        for cell in GetAvailableCells(position):
            position[cell - 1] = "O"

            Eval = minimax(position, depth + 1, alpha, beta, True)
            minEval = min(minEval, Eval)
            beta = min(beta, Eval)
            position[cell - 1] = cell
            if beta <= alpha:
                break
        return minEval
````


1.   FindBestMove :
Duyệt tất cả ô trống, đánh giá bằng Minimax, chọn ô có giá trị tốt nhất.
MAX (X) chọn giá trị lớn nhất, MIN (O) chọn giá trị nhỏ nhất.

2.   main():
Điều khiển vòng lặp trò chơi, cho phép AI và người chơi đi luân phiên.
In bàn cờ, nhập nước đi, kiểm tra thắng/thua hoặc hòa.


3.  currentGame :
Danh sách 1D 9 ô, lưu trạng thái bàn cờ.
Mỗi ô chứa số (trống) hoặc "X"/"O".



4.   counter:
Đếm số lượt đi để xác định hòa khi bàn cờ đầy.

5.   Cơ chế AI:
Sử dụng thuật toán Minimax với Alpha-Beta để chọn nước đi tối ưu.
Luôn đảm bảo chiến thắng hoặc hòa nếu AI chơi hoàn hảo.

````python
import math

def FindBestMove(currentPosition, AI):
    bestVal = -math.inf if AI == "X" else +math.inf
    bestMove = -1
    for cell in GetAvailableCells(currentPosition):
        currentPosition[cell - 1] = AI
        moveVal = minimax(
            currentPosition, 0, -math.inf, +math.inf,
            False if AI == "X" else True
        )
        currentPosition[cell - 1] = cell
        if (AI == "X" and moveVal > bestVal) or (AI == "O" and moveVal < bestVal):
            bestMove = cell
            bestVal = moveVal
    return bestMove


def main():
    player = input("Play as X or O? ").strip().upper()
    AI = "O" if player == "X" else "X"
    currentGame = [*range(1, 10)]
    currentTurn = "X"
    counter = 0
    while True:
        if currentTurn == AI:
            cell = FindBestMove(currentGame, AI)
            currentGame[cell - 1] = AI
            currentTurn = player
        elif currentTurn == player:
            PrintBoard(currentGame)
            while True:
                try:
                    humanInput = int(input("Enter Number (1-9): ").strip())
                    if humanInput in currentGame:
                        currentGame[humanInput - 1] = player
                        currentTurn = AI
                        break
                    else:
                        PrintBoard(currentGame)
                        print("Cell Not Available.")
                except ValueError:
                    print("Invalid input! Enter a number between 1-9.")
        winner = GetWinner(currentGame)
        if winner != None:
            PrintBoard(currentGame)
            print(f"{winner} WON!!!")
            break

        counter += 1
        if winner == None and counter == 9:
            PrintBoard(currentGame)
            print("Tie.")
            break
if __name__ == "__main__":
    main()
````

Đường chéo phụ: từ ô trên cùng bên phải đến ô dưới cùng bên trái.

Chỉ số ô theo hàng i là i*n + (n-1-i).

Nếu tất cả ô trong chéo giống nhau và không trống → trả về người thắng.

````python

def get_winner(board, n):
    """Kiểm tra tất cả hàng, cột, đường chéo"""
    # Hàng
    for i in range(n):
        row = board[i*n:(i+1)*n]
        if all(cell == row[0] for cell in row) and row[0] in [X, O]:
            return row[0]

    # Cột
    for j in range(n):
        col = [board[i*n + j] for i in range(n)]
        if all(cell == col[0] for cell in col) and col[0] in [X, O]:
            return col[0]

    # Chéo chính
    diag1 = [board[i*n + i] for i in range(n)]
    if all(cell == diag1[0] for cell in diag1) and diag1[0] in [X, O]:
        return diag1[0]

    # Chéo phụ
    diag2 = [board[i*n + (n-1-i)] for i in range(n)]
    if all(cell == diag2[0] for cell in diag2) and diag2[0] in [X, O]:
        return diag2[0]

    return None
````

````python
# -----------------------------
# Minimax với Alpha-Beta
# -----------------------------
def minimax(position, depth, alpha, beta, is_maximizing, n, AI, Human):
    winner = get_winner(position, n)
    if winner == AI:
        return 10 - depth
    elif winner == Human:
        return depth - 10
    elif not get_available_cells(position):
        return 0

    if is_maximizing:
        max_eval = -math.inf
        for cell in get_available_cells(position):
            position[cell - 1] = AI
            eval = minimax(position, depth + 1, alpha, beta, False, n, AI, Human)
            position[cell - 1] = cell
            max_eval = max(max_eval, eval)
            alpha = max(alpha, eval)
            if beta <= alpha:
                break
        return max_eval
    else:
        min_eval = math.inf
        for cell in get_available_cells(position):
            position[cell - 1] = Human
            eval = minimax(position, depth + 1, alpha, beta, True, n, AI, Human)
            position[cell - 1] = cell
            min_eval = min(min_eval, eval)
            beta = min(beta, eval)
            if beta <= alpha:
                break
        return min_eval
````

````python

def find_best_move(board, n, AI, Human):
    best_val = -math.inf if AI == X else math.inf
    best_move = -1
    for cell in get_available_cells(board):
        board[cell - 1] = AI
        move_val = minimax(board, 0, -math.inf, math.inf, False if AI == X else True, n, AI, Human)
        board[cell - 1] = cell
        if (AI == X and move_val > best_val) or (AI == O and move_val < best_val):
            best_val = move_val
            best_move = cell
    return best_move
````

Game kết thúc khi có người thắng hoặc bàn đầy (hòa).

Không cho phép AI vs AI.

````python


def main():
    # Chọn kích thước bàn cờ
    while True:
        try:
            n = int(input("Nhập kích thước bàn cờ N (>=3): "))
            if n >= 3:
                break
            else:
                print("Bàn cờ phải >= 3x3!")
        except ValueError:
            print("Nhập số nguyên hợp lệ!")

    # Chọn chế độ: 1) Người vs AI  2) Người vs Người
    while True:
        mode = input("Chọn chế độ: 1) Người vs AI  2) Người vs Người: ").strip()
        if mode in ["1", "2"]:
            break
        print("Nhập 1 hoặc 2!")

    board = initial_state(n)
    current_turn = X
    move_count = 0
    max_moves = n * n

    if mode == "1":
        # Người vs AI
        while True:
            # Chọn ký hiệu người chơi
            while True:
                Human = input("Chọn X hoặc O: ").strip().upper()
                if Human in [X, O]:
                    break
                print("Sai ký hiệu! Chỉ X hoặc O.")
            AI = O if Human == X else X
            break

        while True:
            print_board(board, n)
            winner = get_winner(board, n)
            if winner:
                print(f"Game Over! {winner} thắng!")
                break
            elif move_count == max_moves:
                print("Hòa!")
                break

            if current_turn == AI:
                print(f"AI ({AI}) đang đi...")
                move = find_best_move(board, n, AI, Human)
                board[move - 1] = AI
                current_turn = Human
            else:
                while True:
                    try:
                        user_move = int(input(f"Nhập vị trí (1-{max_moves}): "))
                        if user_move in board and board[user_move - 1] not in [X, O]:
                            board[user_move - 1] = Human
                            current_turn = AI
                            break
                        else:
                            print("Ô không hợp lệ hoặc đã được đánh!")
                    except ValueError:
                        print("Phải nhập số nguyên hợp lệ!")

            move_count += 1

    else:
        # Người vs Người
        players = [X, O]
        while True:
            print_board(board, n)
            winner = get_winner(board, n)
            if winner:
                print(f"Game Over! {winner} thắng!")
                break
            elif move_count == max_moves:
                print("Hòa!")
                break

            current_player = players[move_count % 2]
            while True:
                try:
                    user_move = int(input(f"Người {current_player}, nhập vị trí (1-{max_moves}): "))
                    if user_move in board and board[user_move - 1] not in [X, O]:
                        board[user_move - 1] = current_player
                        break
                    else:
                        print("Ô không hợp lệ hoặc đã được đánh!")
                except ValueError:
                    print("Phải nhập số nguyên hợp lệ!")

            move_count += 1

if __name__ == "__main__":
    main()
````
