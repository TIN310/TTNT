GetWinner(board): Hàm này nhận một tham số board, có thể là một danh sách (list) hoặc mảng (array) đại diện cho 9 ô trên bàn cờ Tic-Tac-Toe. Mỗi phần tử trong danh sách này đại diện cho một ô trên bàn cờ.

1.   Kiểm tra các hàng ngang: Các điều kiện board[0] == board[1] == board[2], board[3] == board[4] == board[5], và board[6] == board[7] == board[8] kiểm tra xem có ba ký hiệu giống nhau liên tiếp trên bất kỳ hàng nào không. Điều kiện and board[x] != " " đảm bảo rằng đó không phải là các ô trống.
2.   Kiểm tra các cột dọc: Tương tự, các điều kiện board[0] == board[3] == board[6], board[1] == board[4] == board[7], và board[2] == board[5] == board[8] kiểm tra xem có ba ký hiệu giống nhau liên tiếp trên bất kỳ cột nào không.




```python
import os , math
def GetWinner(board) :
    if board[0] == board[1] == board[2] and board[0] != " ":
        return board[0]
    elif board[3] == board[4] == board[5] and board[3] != " ":
        return board[3]
    elif board[6] == board[7] == board[8] and board[6] != " ":
        return board[6]
    elif board[0] == board[3] == board[6] and board[0] != " ":
        return board[0]
    elif board[1] == board[4] == board[7] and board[1] != " ":
        return board[1]
    elif board[2] == board[5] == board[8] and board[2] != " ":
        return board[2]
    elif board[0] == board[4] == board[8] and board[0] != " ":
        return board[0]
    elif board[2] == board[4] == board[6] and board[2] != " ":
        return board[2]
    return None
```

PrintBoard(board) là hàm hiển thị bàn cờ Tic-Tac-Toe ra màn hình, giúp người chơi hoặc AI dễ nhìn thấy trạng thái game hiện tại.

GetAvailableCells(board) là hàm xác định tất cả các ô mà người chơi hoặc AI có thể đi tiếp.


1.   available = list():
Tạo một danh sách rỗng để lưu các ô còn trống trên bàn cờ.
2.  for cell in board:
Duyệt qua tất cả các ô (0 → 8 hoặc ký hiệu trong danh sách board).


1.   if cell != "X" and cell != "O":
Kiểm tra ô hiện tại có bị đánh chưa.
Nếu chưa → là ô trống.
2.   available.append(cell):
Thêm ô trống vào danh sách available.


1.   return available:
Trả về danh sách tất cả các ô còn trống, dùng cho AI hoặc người chơi chọn nước đi.




```python
def GetAvailableCells(board):
    available = list()
    for cell in board:
        if cell != "X" and cell != "O":
            available.append(cell)
    return available
```

Hàm minimax được triển khai nhằm tìm nước đi tối ưu cho AI trong trò chơi Tic-Tac-Toe.
Thuật toán giả định rằng cả AI và đối thủ đều chơi tối ưu, và lựa chọn nước đi dựa trên việc tối đa hóa lợi ích cho AI hoặc tối thiểu hóa lợi ích cho đối thủ.

**Tham số**

position :	Danh sách 1D 9 ô biểu diễn trạng thái bàn cờ hiện tại.

depth :	Độ sâu hiện tại trong cây tìm kiếm. Dùng để ưu tiên chiến thắng sớm và kéo dài thất bại.

alpha :	Giá trị tốt nhất mà người chơi MAX có thể đảm bảo tại nhánh hiện tại.

beta :	Giá trị tốt nhất mà người chơi MIN có thể đảm bảo tại nhánh hiện tại.

isMaximizing :True nếu lượt MAX (thường là X), False nếu lượt MIN (thường là O).
Kiểm tra trạng thái kết thúc (Base Case):

 **Nguyên lý hoạt động**

Nếu X thắng → trả về 10 - depth.

Nếu O thắng → trả về -10 + depth.

Nếu hòa → trả về 0.

Ý nghĩa: depth được trừ/cộng để ưu tiên chiến thắng nhanh hoặc kéo dài thất bại.

Lượt MAX (X):

Khởi tạo giá trị maxEval = -∞.

Thử tất cả các ô còn trống, đặt X vào ô và gọi đệ quy Minimax cho lượt MIN.

Cập nhật maxEval = max(maxEval, Eval) và alpha = max(alpha, Eval).

Hoàn tác nước đi (backtracking).

Nếu beta <= alpha → cắt tỉa nhánh (Alpha-Beta pruning).

Lượt MIN (O):

Khởi tạo giá trị minEval = +∞.

Thử tất cả các ô còn trống, đặt O vào ô và gọi đệ quy Minimax cho lượt MAX.

Cập nhật minEval = min(minEval, Eval) và beta = min(beta, Eval).

Hoàn tác nước đi (backtracking).

Nếu beta <= alpha → cắt nhánh không cần thiết.

```python

def minimax(position, depth, alpha, beta, isMaximizing):
    winner = GetWinner(position)
    if winner != None:
        return 10 - depth if winner == "X" else -10 + depth
    if len(GetAvailableCells(position)) == 0:
        return 0
    if isMaximizing:
        maxEval = -math.inf
        for cell in GetAvailableCells(position):
            position[cell - 1] = "X"
            Eval = minimax(position, depth + 1, alpha, beta, False)
            maxEval = max(maxEval, Eval)
            alpha = max(alpha, Eval)
            position[cell - 1] = cell
            if beta <= alpha:
                break
        return maxEval
    else:
        minEval = +math.inf
        for cell in GetAvailableCells(position):
            position[cell - 1] = "O"

            Eval = minimax(position, depth + 1, alpha, beta, True)
            minEval = min(minEval, Eval)
            beta = min(beta, Eval)
            position[cell - 1] = cell
            if beta <= alpha:
                break
        return minEval
```


1.   FindBestMove :
Duyệt tất cả ô trống, đánh giá bằng Minimax, chọn ô có giá trị tốt nhất.
MAX (X) chọn giá trị lớn nhất, MIN (O) chọn giá trị nhỏ nhất.

2.   main():
Điều khiển vòng lặp trò chơi, cho phép AI và người chơi đi luân phiên.
In bàn cờ, nhập nước đi, kiểm tra thắng/thua hoặc hòa.


3.  currentGame :
Danh sách 1D 9 ô, lưu trạng thái bàn cờ.
Mỗi ô chứa số (trống) hoặc "X"/"O".



4.   counter:
Đếm số lượt đi để xác định hòa khi bàn cờ đầy.

5.   Cơ chế AI:
Sử dụng thuật toán Minimax với Alpha-Beta để chọn nước đi tối ưu.
Luôn đảm bảo chiến thắng hoặc hòa nếu AI chơi hoàn hảo.

```python
import math

def FindBestMove(currentPosition, AI):
    bestVal = -math.inf if AI == "X" else +math.inf
    bestMove = -1
    for cell in GetAvailableCells(currentPosition):
        currentPosition[cell - 1] = AI
        moveVal = minimax(
            currentPosition, 0, -math.inf, +math.inf,
            False if AI == "X" else True
        )
        currentPosition[cell - 1] = cell
        if (AI == "X" and moveVal > bestVal) or (AI == "O" and moveVal < bestVal):
            bestMove = cell
            bestVal = moveVal
    return bestMove


def main():
    player = input("Play as X or O? ").strip().upper()
    AI = "O" if player == "X" else "X"
    currentGame = [*range(1, 10)]
    currentTurn = "X"
    counter = 0
    while True:
        if currentTurn == AI:
            cell = FindBestMove(currentGame, AI)
            currentGame[cell - 1] = AI
            currentTurn = player
        elif currentTurn == player:
            PrintBoard(currentGame)
            while True:
                try:
                    humanInput = int(input("Enter Number (1-9): ").strip())
                    if humanInput in currentGame:
                        currentGame[humanInput - 1] = player
                        currentTurn = AI
                        break
                    else:
                        PrintBoard(currentGame)
                        print("Cell Not Available.")
                except ValueError:
                    print("Invalid input! Enter a number between 1-9.")
        winner = GetWinner(currentGame)
        if winner != None:
            PrintBoard(currentGame)
            print(f"{winner} WON!!!")
            break

        counter += 1
        if winner == None and counter == 9:
            PrintBoard(currentGame)
            print("Tie.")
            break
if __name__ == "__main__":
    main()
```
